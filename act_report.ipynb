## Report: act_report

Real-world data don't always come clean, most often than not they're messy or dirty, and sometimes messy and dirty all at the same time. This is the more reason why we need to clean our data before we analyse and visualize it.

In the instant project, I wrangled, analized, and visualized the tweet archive of Twitter user @dog_rates, also known as WeRateDogs. In this project, I worked on the following three datasets.

1. Enhanced Twitter Archive
2. Additional Data via the Twitter API
3. Image Predictions File

### The following are my analysis and insights:

#### 1. Explore the data description for better insights on the master dataset
After cleaning the data, I explored it even further,and I found the following 
insights from my analysis

1. The p1-p3_conf columns shouldn't be greater than one because they are confidence intervals. As a result, the values range from 0 to 1.

2. The data shows that tweets had more engagements than retweets

3. The post with the least image had one image while others with the highest image had about 4 images

4. The differnce in percentage from 75% to the max is very high, this is based on the fact that some tweets had outrageous rating, and the min with 0 is as a result of tweets without images.

#### 2. What is the reason behind predicting dog breeds with low and high p1 conf.?

We saw from earlier that the data contains invalid predictions. Some are actually dogs, but they have low p1_conf. exploring the dataset further we realized that the reason behind the wrong prediction is as a result of the noise in the image.
